# Heatmap

### Heatmap is used to visualize 2-D data like a matrix or table using colors

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights_df = sns.load_dataset('flights').pivot('month', 'year', 'passengers')
```

```python
flights_df
```

flights_df is a matrix with 1 row as month and 1 column for each year, values showed is the number of passengers in thousands that visited the airport in a specific month of a year.
We can use sns.heatmap() function to visualize the footfall of the airport

```python
plt.title("No. Of Passengers (1000s)")
sns.heatmap(flights_df);
```

### By visualizing above heatmap, we can conclude:
* the most number of footfalls happened in July and August month
* footfalls has increased year by year for each months in each year

```python
plt.title("Number of Passengers at Airport in 1000s")
sns.heatmap(flights_df,fmt='d', annot=True, cmap='Greens');
```
