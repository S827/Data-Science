### customizing x axis

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(years, yield_apples);
```

## add the axis labels with plt.xlabel() and plt.ylabel()

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)');
```

## Plotting multiple lines

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, oranges)
plt.plot(years, apples)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)');
```

## x and y axis changed

```python
plt.plot(oranges,  years)
plt.plot(apples, years)
plt.ylabel('Year')
plt.xlabel('Yield(tons per hectare)');
```

## Chart title and legend

title with plt.title() and legend with plt.legend()

```python
plt.plot(years, oranges)
plt.plot(years, apples)

plt.xlabel('year')
plt.ylabel('yields')
plt.title('Crop yields in Kanto region')
plt.legend(['oranges', 'apples']);
```

## Line markers

```python
plt.plot(years, apples, marker="^")
plt.plot(years, oranges, marker="v")
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Crop Yields in Kanto region')
plt.legend(['Apple', 'Orange']);
```

## Styling lines and markers

* color- sets the color of the line
* marker - sets the marker symbol
* ls - sets the line style
* lw - set the line width
* ms - sets the marker size
* mec - sets the marker edge color
* mew - sets the marker edge width size
* mfc - sets the marker face color
* alpha - sets the opacity of plot

```python
plt.plot(years, oranges, color='red', marker='v', ls='--', lw=5, ms=15, mec='blue', mew=3, mfc='white', alpha=0.8)
plt.plot(years, apples, color='green', marker='^', ls='-', lw=5, ms=15, mec='blue', mew=3, mfc='black', alpha=0.5)

plt.xlabel('year')
plt.ylabel('yields')
plt.title('Crop yields in Kanto region')
plt.legend(['oranges', 'apples']);
```

## fmt argument provides shorthand for marker shape, line style, and line color, can only be intoduced as an 3rd argument in the plt.plot argument

```python
plt.plot(years, apples, '^-g')
plt.plot(years, oranges, 'v--r')
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Crop Yields in Kanto Region')
plt.legend(['Apples', 'Oranges']);
```

## Changing the figure size

### use plt.figure() function to alter the plot size, arguments figsize=() 

```python
plt.figure(figsize=(5, 5))
plt.plot(years, apples, '^-g')
plt.xlabel('Year')
plt.ylabel('Yields')
plt.title('Yields in XYZ')
plt.legend(['Apple']);
```

```python
plt.figure(figsize=(10, 5))
plt.plot(years, apples, '^-g')
plt.xlabel('Year')
plt.ylabel('Yields')
plt.title('Yields in XYZ')
plt.legend(['Apple']);
```

# improving default styles using seaborn library

## sns.set_style()

### sns.set_style(whitegrid)

```python
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(years, apples, '^-g')
plt.xlabel('Year')
plt.ylabel('Yields')
plt.title('Yields in XYZ')
plt.legend(['Apple']);
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
plt.plot(years, apples, '^-g')
plt.xlabel('Year')
plt.ylabel('Yields')
plt.title('Yields in XYZ')
plt.legend(['Apple']);
```
