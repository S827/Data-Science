# Scatter Plot

```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris
```

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

## we will try to visualize the relationship between sepal length and sepal width

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(8, 5))
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

Above plot is not very informative, we can use scatter plot to visualize the relationship between sepal length and sepal width using the scatterplot function from seaborn module

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=flowers_df.sepal_length, y = flowers_df.sepal_width);
```

## we can add colors as per species  with hue argument and enlarge the points with s argument

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x = flowers_df.sepal_length, y = flowers_df.sepal_width, hue = flowers_df.species, s = 100);
```

With above scatterplot we can conclude by just visualizing at it
* setosa flowers have smaller length but higher sepal widths
* versicolor flowers have lesser sepal width than virginica and setosa, and sepal length is greater than setosa but less than virginica
* virginica flowers have lesser sepal width than setosa and longer width than versicolor, and sepal lengths are greater than other 2 species

## customizing Seaborn figures

```python
plt.figure(figsize=(10, 6))
plt.title('Sepal Dimensions')
sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width,
                hue=flowers_df.species,
                s=100);
```

## Plotting using pandas Dataframse

```python
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length',
                y='sepal_width',
                hue='species',
                s=100,
                data=flowers_df);
```
