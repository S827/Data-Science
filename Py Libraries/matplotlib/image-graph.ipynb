# Images in graph

matplotlib can be used to display images in a graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from urllib.request import urlretrieve
```

```python
urlretrieve('https://i.imgur.com/SkPbq.jpg', 'chart.jpg')
```

## to display image, it needs to be read by PIL module

```python
from PIL import Image
```

```python
img = Image.open('chart.jpg')
```

```python
img
```

### An image laoded using PIL module is simply a 3-D numpy array, containing pixel intensities of RGBof the image, we can covert the image into an array using np.array

```python
import numpy as np
```

```python
img_array = np.array(img)
```

```python
img_array
```

```python
img_array.shape
```

## We can display image using plt.imshow(img)

```python
plt.imshow(img);
```

```python
#turn off the grid
plt.grid(False)

#give it a title
plt.title('A data science meme')

#disable the axis values with plt.axis() function
plt.axis('off')

#display the image
plt.imshow(img);
```

### To display the slice of the image, slice the numpy array

```python
#turn off the grid
plt.grid(False)

#turn off axis
plt.axis('off')

#display image with sliced elements
plt.imshow(img_array[125:325, 105:305])
```
