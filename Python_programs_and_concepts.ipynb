# 1. Take input of date from user in mm dd yyyy format and output the Day name 

```python
# import calendar module
import calendar

# find_Day function
def find_day(date):
#     convert the date into integer values and store into variables using map
    month, day, year = map(int, date.split())
#     print(month, day, year)
#     get the day from the integer values of date using calendar.weekday function
    day_int = calendar.weekday(year, month, day)
    print(day_int)
#     convert the day integer into name of the day using calendar.day_name function
    day_name = calendar.day_name[day_int]
    return day_name.upper()
# main func
if __name__ == "__main__":
    # get the date as input
    date = input()
#     call find_day function
    weekday = find_day(date)
# print day
    print(weekday)
```

# 3. There are  students in this class whose names and grades are assembled to build the following list:

python students = [['Harry', 37.21], ['Berry', 37.21], ['Tina', 37.2], ['Akriti', 41], ['Harsh', 39]]

The lowest grade of  belongs to Tina. The second lowest grade of  belongs to both Harry and Berry, so we order their names alphabetically and print each name on a new line.

```python
def find_2nd_low_stud(students):
    # extract the grades
    grades = [grade for name, grade in students]
    # find unique grades and sort them
    unique_grade = sorted(set(grades))
    # get the second lowest grade
    if len(unique_grade) < 2:
        return []  # return empty list if there are less than 2 grades
    second_lowest_grade = unique_grade[1]
    # get the name of student who got second lowest grade
    second_lowest_stud = [name for name, grade in students if grade == second_lowest_grade]
    # sort the student names alphabetically
    second_lowest_stud.sort()
    return second_lowest_stud

def print_students(second_lowest_stud):
    print(second_lowest_stud)
if __name__ == "__main__":
    students = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        students.append([name, score])
    second_lowest_student = find_2nd_low_stud(students)
    print_students(second_lowest_student)
```

```python
def calcPoly(val, poly):
    #     replace x with val
    poly_str = poly.replace('x', str(val))
    # calcualte polynomial with eval function
    calc = eval(poly_str)
    return calc

if __name__ == "__main__":
    val, out = map(int, input().split())
    poly = input()
    output = calcPoly(val, poly)
    print(out == output)
```

```python
def command(N):
    i = 0
    ls = []
    while i < N:
        com = input()
        parts = com.split()
        if len(parts) == 1:
            if com == 'print':
                print(ls)
            elif com == 'sort':
                ls.sort()
            elif com == 'pop':
                ls.pop()
            elif com == 'reverse':
                ls.reverse()
        elif len(parts) == 2:
            if parts[0] == 'remove':
                ls.remove(int(parts[1]))
            elif parts[0] == 'append':
                ls.append(int(parts[1]))
        elif len(parts) == 3:
            if parts[0] == 'insert':
                ls.insert(int(parts[1]), int(parts[2]))
        i += 1
if __name__ == '__main__':
    N = int(input())
    command(N)
```

# 8. Swap Case

```python
def swapcase(s):
    a = []
    for i in range(len(s)):
        if s[i].isupper():
            a.append(s[i].lower())
        elif s[i].islower():
            a.append(s[i].upper())
        else:
            a.append(s[i])
    return ''.join(a)

if __name__ == "__main__":
    s = input()
    res = swapcase(s)
    print(res)
```

# 9. Split and Join in Strings

```python
def splitJoin(a):
    ls = a.split(" ")
    return "-".join(ls)

if __name__ == "__main__":
    a = input()
    res = splitJoin(a)
    print(res)
```
